```python
import pandas as pd
import os
```

```python
from os.path import join as opj
root_path = "/Users/abeukers/wd/coffeworld_exp/csw_k"
df_dir = opj(root_path,'dataframes')

DF_list = []
for fname in os.listdir(df_dir)[1:]:
  fpath = opj(root_path,"dataframes",fname)
  sid = fname.split('-')[-1].split('.')[0]
  df = pd.DataFrame.from_csv(fpath)
  df['global_trial_index'] = df.index + len(df)*int(sid)
  DF_list.append(df)
  
full_df = pd.concat(DF_list)
  
```
